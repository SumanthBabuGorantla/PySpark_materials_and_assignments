```
#/FileStore/tables/Airplane_dataset.csv
#a.) Write code that to read the csv File?

df=spark.read.option("header","True").option("inferSchema","True").csv("/FileStore/tables/Airplane_dataset.csv")
df.show()
```

```
#b.) How many unique airlines are present in this dataset?

df.select("UniqueCarrier").distinct().show()
df.select("UniqueCarrier").distinct().count()
```

```
# c.) Which airline is delayed on departure most often ?
#DepDelay,UniqueCarrier
from pyspark.sql.functions import *
df.filter(col("IsDepDelayed")=="YES").groupBy("UniqueCarrier").agg(count("ISDepDelayed").alias("count_of_DepDelayed")).show(1)
```

```
#d.) What was the average arrival delay per airline?
from pyspark.sql.functions import *
df.filter(col("IsArrDelayed")=="YES").groupBy("UniqueCarrier").agg(round(avg("ArrDelay"),2)).show()
```
