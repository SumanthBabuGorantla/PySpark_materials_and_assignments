```
#/FileStore/tables/UIDAI_ENR_DETAIL_20170308.csv


df=spark.read.option("inferSchema","True").option("header","True").csv("/FileStore/tables/UIDAI_ENR_DETAIL_20170308.csv")
df.show(truncate=False)
```

```
import pyspark.sql.functions
from pyspark.sql.functions import *
```

```
#1) Create a dataframe with Total Aadhaar's generated for each state

Total_Adhar_per_State=df.groupBy("State").agg(sum("Aadhaar generated").alias ("Total_Adhar_per_State"))
Total_Adhar_per_State.show()
```

```
# 2) Create a dataframe with Total Aadhaar's generated by each enrollment agency

Adhars_per_Enrollment_Agency=df.groupBy("Enrolment Agency").agg(sum("Aadhaar generated").alias("Adhars_per_Enrollment_Agency"))
Adhars_per_Enrollment_Agency.show()
```

```
# 3) Create dataframe with top 10 districts with maximum Aadhaar's generated for both Male and Female?

Male_Female_Df=df.filter(df.Gender!="T")
#Male_Female_Df.show()
Male_Female_Df=Male_Female_Df.groupBy("District").agg(sum("Aadhaar generated").alias("Total_Aadhaar_M_F")).orderBy(col("Total_Aadhaar_M_F").desc())
Male_Female_Df.show(10)
```

```
# 4) Create a dataframe with Total Aadhaar's generated for top 10 least states
Adhar_Per_State=df.groupBy("State").agg(sum("Aadhaar generated").alias("Adhar_Per_State")).orderBy("Adhar_Per_State")
Adhar_Per_State.show(10)
```

```
# 5) for which age most adhar card has declined ?
Declined_Aadhaar=df.groupBy("Age").agg(sum("Enrolment Rejected").alias("Declined_Aadhaar")).orderBy(col("Declined_Aadhaar").desc())
Declined_Aadhaar.show(1)
```
