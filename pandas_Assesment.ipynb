```
"""2. Create a function that takes in two lists named keys and values as arguments. 
Keys would be strings and contain n string values. 
Values would be a list containing n lists. 
The methods should return a new pandas dataframe with keys as column names and values as 
their corresponding values 
e.g. -> create_dataframe(["One", "Two"], [["X", "Y"], ["A", "B"]]) -> should return a dataframe 
 
One Two 
 
0 X A 
 
1 Y B"""
import pandas as pd

def create_dataframe(keys, values):
    dict_1 = {}
    for i in range(len(keys)):
        dict_1[keys[i]] = values[i]
    df = pd.DataFrame(dict_1)
    return df
create_dataframe(["One", "Two"], [["X", "Y"], ["A", "B"]])
```

```
import pandas as pd
#3.Create a function that concatenates two dataframes. Use previously created function to create two dataframes and pass them as parameters Make sure that the indexes are reset before returning: 
def create_dataframe1(keys, values):
    dict_1 = {}
    for i in range(len(keys)):
        dict_1[keys[i]] = values[i]
    df1 = pd.DataFrame(dict_1)
    return df1
create_dataframe1(["One", "Two"], [["X", "Y"], ["A", "B"]])


def create_dataframe2(keys, values):
    dict_1 = {}
    for i in range(len(keys)):
        dict_1[keys[i]] = values[i]
    df2 = pd.DataFrame(dict_1)
    return df2
create_dataframe2(["One", "Two"], [["X", "Y"], ["A", "B"]])




def concat_df(df1, df2):
    df = pd.concat([df1, df2], axis=0, ignore_index = True)
    return df

concat_df(create_dataframe1(["One", "Two"], [["X", "Y"], ["A", "B"]]),create_dataframe2(["One", "Two"], [["X", "Y"], ["A", "B"]]))
```
