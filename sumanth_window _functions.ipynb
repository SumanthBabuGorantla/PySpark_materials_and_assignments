```
from pyspark.sql.functions import *
rdd1 = sc.parallelize([("James", "Sales", 3000),
    ("Michael", "Sales", 4600),
    ("Robert", "Sales", 4100),
    ("Maria", "Finance", 3000),
    ("James", "Sales", 3000),
    ("Scott", "Finance", 3300),
    ("Jen", "Finance", 3900),
    ("Jeff", "Marketing", 3000),
    ("Kumar", "Marketing", 2000),
    ("Saif", "Sales", 4100)])
df=rdd1.toDF(["employee_name", "department", "salary"])
#rdd1.collect()
df.show()
```

```
from pyspark.sql.window import Window
from pyspark.sql.functions import *
windowSpec = Window.partitionBy("department").orderBy("salary")
res=df.withColumn("row_number",row_number().over(windowSpec))
res.show()
```

```
from pyspark.sql.window import Window
from pyspark.sql.functions import *
windwoSpec=Window.partitionBy("department").orderBy("salary")
res=df.withColumn("rank",rank().over(windowSpec))
res.show()
```

```
from pyspark.sql.window import Window
from pyspark.sql.functions import *
windowSpec=Window.partitionBy("department").orderBy("salary")
res=df.withColumn("dense_rank",dense_rank().over(windowSpec))
res.show()
```

```
from pyspark.sql.window import Window
from pyspark.sql.functions import *
windowSpec=Window.partitionBy("Department").orderBy("salary")
res=df.withColumn("lag",lag("salary",1).over(windowSpec))
res.show()
```

```
from pyspark.sql.window import Window
from pyspark.sql.functions import *
windowSpec=Window.partitionBy("department").orderBy("salary")
res=df.withColumn("lead",lead("salary",1).over(windowSpec))
res.show()

```
