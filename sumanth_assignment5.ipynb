```
df1=spark.read.format("com.crealytics.spark.excel").option("header", "true").option("inferSchema","true").load('/FileStore/tables/Gapminder_key_1.xlsx')
df1.show()
```

```
df2=spark.read.format("com.crealytics.spark.excel").option("header", "true").option("inferSchema","true").load('/FileStore/tables/Gapminder_key_2.xlsx')
df2.show()
```

```
df3=df1.join(df2,(df1.year ==  df2.year) & (df1.country==df2.country),"inner")
df3=df3.drop(df1["country"])
df3=df3.drop(df1["year"])
df3.show()
```

```
df3.dtypes
```

```
# Q1: Country and continent are string data types. Please convert them into categorical data types using programming constructs.

from pyspark.sql.functions import *
df3['continent'] = df3.continent.astype('categorical')
df3.types
```

```
#Q2: Count the number of countries in this dataset.

df3.select("country").distinct().count()
```

```
#Q3: Which country had the lowest life expectancy in 2007?
from pyspark.sql.functions import *

df3.filter(col("year")==2007).select("country","life_exp").orderBy((col("life_exp"))).show(1)
```

```
#Q4: Which country had the highest GDP per capita in 2007?
df3.filter(col("year")==2007).select("country","gdp_cap").orderBy(col("gdp_cap").desc()).show(1)
```

```
# Q5: Which country had the highest GDP in 2007? GDP is calculated by multiplying the population into GDP per capita.
df3.filter(col("year")==2007).withColumn("GDP",col("population")*col("gdp_cap")).orderBy(col("GDP").desc()).show(1)
```

```
# Q6: Is there any change in the trend from 1952 to 2007 in life expectancy?

#df3.show()

df3.groupBy("year").agg(avg("life_exp")).show()
```

```
#Q7: Which continent has the lowest life expectancy in 2007? Please use data visualization techniques to answer this question.


df4=df3.filter(col("year")==2007).groupBy("continent").agg(avg("life_exp").alias("avg_life_exp")).orderBy("avg_life_exp")
df4.show()
display(df4)
```

```
# Q8: What is the trend of life expectancy in India from 1952 - 2007?

df3.filter(col("country")=="India").groupBy("year").agg(avg("life_exp").alias("avg_life_exp")).orderBy("avg_life_exp").show()
```

```
# Q9: What is the trend of GDP per capita in India from 1952 - 2007?

df3.filter(col("country")=="India").select("country","year","gdp_cap").show()
```

```
# Q10: What is the median age of each continent by year?
from pyspark.sql.functions import *
df3.groupBy("year","continent").agg(percentile_approx("life_exp", 0.5).alias("median")).show()
```
