```

def get_value(val1):
    if(val1>3):
        return "Greater than 3"
    elif(val1==3):
        return "Equals to 3"
    else:
        return "Lesser than 3"

from pyspark.sql.functions import udf,col
df=spark.read.csv("/FileStore/tables/first_test__1_.csv",header="True",inferSchema="True")

first_udf = udf(get_value) # here the udf function will take the "get_value" as Function
df = df.withColumn("get_res",first_udf(col("id")))
df.show()
```

```
def length(word):
    len1=len(word)
    return len1

from pyspark.sql.functions import udf,col
df=spark.read.csv("/FileStore/tables/first_test__1_.csv",header="True",inferSchema="True")
second_udf = udf(length)
df=df.withColumn("length",second_udf(col("name")))
df.show()
```

```
def length(name,zip):
    len1=len(name)
    len2=len(str(zip))
    return len1+len2
from pyspark.sql.functions import udf,col
df=spark.read.csv("/FileStore/tables/first_test__1_.csv",header="True",inferSchema="True")

third_udf = udf(length)
df=df.withColumn("length",third_udf(col("name"),col("zip")))
df.show()
```

```
def vowels(name):
    j=""
    for i in name:
        if i.lower() in "aeiou":
            j=j+i
    return j
            

from pyspark.sql.functions import udf,col
df=spark.read.csv("/FileStore/tables/first_test__1_.csv",header="True",inferSchema="True")

fourth_udf = udf(vowels)
df=df.withColumn("vowels",fourth_udf("name"))
df.show()
```
