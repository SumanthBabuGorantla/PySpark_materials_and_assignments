```
data1 =sc.textFile("/FileStore/tables/movies.dat").map(lambda x : x.split("::")[1])
data1.collect()
dict1 = {}
data2 = data1.map(lambda x :int(x[x.rindex("(")+1:x.rindex(")")])).distinct().collect()
for i in data1.collect():
  year = int(i[i.rindex("(")+1:i.rindex(")")])
  if(year in data2):
    if(dict1.__contains__(year)):
      get_value = dict1[year]
      get_value.append(i)
      dict1[year] = get_value
    else:
      dict1[year] = [i]
print(dict1)      
```
