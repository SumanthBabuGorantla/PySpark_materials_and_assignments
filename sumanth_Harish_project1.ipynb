```
df=spark.read.csv("/FileStore/tables/data.txt",sep="|",header="True",inferSchema="True")
df.show(100)
```

```
df.printSchema()
```

```
from pyspark.sql.functions import *

df.groupBy("APP_ID","APP_GROUP","APP_NAME").count().orderBy("APP_ID").show()
```

```
from pyspark.sql.functions import *
df.groupBy("APP_GROUP").agg(countDistinct("MSISDN")).show()
```

```
df.select("APP_GROUP").distinct().orderBy("APP_GROUP").limit(5).show()
```

```
from pyspark.sql.functions import *
df.select("APP_GROUP").distinct().limit(5).show()
```

```
#df.select("APP_ID","APP_GROUP","APP_NAME")

df.groupBy("APP_GROUP","APP_NAME").agg(count("APP_NAME")).limit(10).show()
```
