```
"""Find out how many orders, how many products and how many sellers are in the data.
How many products have been sold at least once? Which is the product contained in more orders?"""
```

```
sales=spark.read.parquet("/FileStore/tables/sales_snappy.parquet")
sales.show()
```

```
sellers=spark.read.parquet("/FileStore/tables/sellers_snappy.parquet")
sellers.show()
```

```
products=spark.read.parquet("/FileStore/tables/products_snappy.parquet")
products.show(100)
```

```
#1) Find out how many orders, how many products and how many sellers are in the data.
print("orders count: ",sales.select("order_id").distinct().count())
print("products count: ",products.select("product_id").distinct().count())
print("sellers count: ",sellers.select("seller_id").distinct().count())



```

```
#2) How many products have been sold at least once? Which is the product contained in more orders?

products_sales=products.join(sales,products.product_id ==  sales.product_id,"inner")
products_sales=products_sales.drop(sales["product_id"])
products_sales.show()
```

```
#2) How many products have been sold at least once? 

from pyspark.sql.functions import *
products_sales.groupBy(col("product_name")).count().orderBy(desc("count")).show()
```

```
#Which is the product contained in more orders?

products_sales.groupBy("product_name").count().orderBy(desc("count")).limit(1).show()
```

```
#How many distinct products have been sold in each day?
products_sales.groupBy("date").agg(countDistinct("product_id")).show()
```

```
#What is the average revenue of the orders?

products_sales.groupBy("order_id").agg(avg("price")).show()
```

```
#For each seller, what is the average % contribution of an order to the seller's daily quota?

sales_seller=sales.join(broadcast(sellers), sales["seller_id"] == sellers["seller_id"], "inner").withColumn(
    "ratio", sales["num_pieces_sold"]/sellers["daily_target"])
#sales_seller.show()
sales_seller=sales_seller.drop(sales["seller_id"])
sales_seller.groupBy("seller_id").agg(avg("ratio")).show()
```
