```
from pyspark.sql.functions import *
rdd1 = sc.parallelize([("James", "Sales", 3000),
    ("Michael", "Sales", 4600),
    ("Robert", "Sales", 4100),
    ("Maria", "Finance", 3000),
    ("James", "Sales", 3000),
    ("Scott", "Finance", 3300),
    ("Jen", "Finance", 3900),
    ("Jeff", "Marketing", 3000),
    ("Kumar", "Marketing", 2000),
    ("Saif", "Sales", 4100)])
emp=rdd1.toDF(["employee_name", "department", "salary"])
#rdd1.collect()
emp.show()
```

```
def salary(salary):
    if salary>3000:
        return "good pay"
    else:
        return "bad pay"


salary_udf=udf(salary)
emp.withColumn("sal_range",salary_udf("salary")).show()
```
