```
#/FileStore/tables/olympix_data_organized_with_header__1_.csv


#Q1: Write PySpark code to read olympix_data.csv file

olympic=spark.read.csv("/FileStore/tables/olympix_data_organized_with_header__1_.csv",header=True,inferSchema=True)
olympic.show()
```

```
#Q2: Write PySpark code to print the Olympic Sports/games in the dataset.


olympic.select("sports").distinct().show()
```

```
olympic.show()
```

```
#Q3: Write PySpark code to plot the total number of medals in  each Olympic Sport/game. Sort the result based on the the total number of medals.
from pyspark.sql.functions import *

df=olympic.groupBy("sports").agg(sum(col("total_medal")).alias("total_medals")).orderBy("total_medals")
display(df)
```

```
#Q4: Find the total number of medals won by each country in swimming.

olympic2=olympic.filter(col("sports")=="wimming")
olympic2.groupBy("country").agg(sum(col("total_medal")).alias("total_medals")).show()
```

```
#Q5: Find the total number of medals won by each country in Skeleton.
olympic2=olympic.filter(col("sports")=="Skeleton")
olympic2.groupBy("country").agg(sum(col("total_medal")).alias("total_medals")).show()
```

```
#Q6 Find the number of medals that the US won yearly.

olympic2=olympic.filter(col("country")=="United States")
olympic2.groupBy("year").agg(sum(col("total_medal")).alias("total_medals")).show()
```

```
#Q7 Find the total number of medals won by each country.


olympic.groupBy("country").agg(sum(col("total_medal")).alias("total_medals")).show()
```

```
#Q8 Who was the oldest athletic in the olympics? whcih country was he/she from?

olympic.select("name","age","country").orderBy(desc("age")).limit(1).show()
```

```
from pyspark.sql.window import Window
from pyspark.sql.functions import *

windowSpec=Window.partitionBy("name").orderBy(desc("age"))
olympic2=olympic.withColumn("Age_Rank",row_number().over(windowSpec))
olympic2.select("name","age","Age_Rank").orderBy(desc("age")).show()
```
