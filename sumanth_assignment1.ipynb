```
data=sc.textFile("/FileStore/tables/user_data.txt")
data=data.map(lambda line:line.split(","))
data.collect()
```

```
#1.How many unique professions do we have in the data file?

unique_profession = data.map(lambda a: a[3])
#unique_profession.collect()
#unique_profession.distinct().collect()
unique_profession.distinct().count()
```

```
#How many different users belongs to unique professions?

data=sc.textFile("/FileStore/tables/user_data.txt")
data=data.map(lambda line:line.split(","))
data.collect()
users_count = data.map(lambda a : (a[3],1)).reduceByKey(lambda x,y:x+y)
users_count.collect()
```

```
#How many users belong to a unique zip code in the sample file ?

data=sc.textFile("/FileStore/tables/user_data.txt")
data=data.map(lambda line:line.split(","))
data.collect()
zipcode_count=data.map(lambda a:(a[4],1)).reduceByKey(lambda x,y:x+y)
zipcode_count.collect()
```

```
# How many users are male and female ?


data=sc.textFile("/FileStore/tables/user_data.txt")
data=data.map(lambda line:line.split(","))
data.collect()
gender=data.map(lambda a:(a[2],1)).reduceByKey(lambda x,y:x+y)
gender.collect()
```
